```python
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import random
import scipy.signal
```

```python
# Step 1: Load RGB image from URL
image_url = "https://2.bp.blogspot.com/-mNeh824oxrk/VkFfAzw7hiI/AAAAAAAAAN4/CsBakwhIlLc/s1600/rgb.png"
image_path, _ = urllib.request.urlretrieve(image_url)
rgb_image = Image.open(image_path)
```

```python
# Step 2: Plot the loaded image
plt.imshow(rgb_image)
plt.title("Original Image")
plt.show()
```

```python
# Step 3: Show the shape of the resulting image array
print("Original Image Shape:", np.array(rgb_image).shape)
```

```python
# Step 4: Resize image to 224x224
resized_image = rgb_image.resize((224, 224))
```

```python
# Step 5: Plot the resized image
plt.imshow(resized_image)
plt.title("Resized Image (224x224)")
plt.show()
```

```python

# Step 6: Show the shape of the resized image array
print("Resized Image Shape:", np.array(resized_image).shape)
```

```python
# Step 7: Show grayscale copy
grayscale_image = resized_image.convert("L")
```

```python

# Step 8: Show the shape of the new grayscale image
print("Grayscale Image Shape:", np.array(grayscale_image).shape)
```

```python
# Step 9: Plot the resulting grayscale image
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image (224x224)")
plt.show()
```

```python
# Step 10: Convolve with random filters
num_filters = 10
filter_size = 3  # You can adjust the filter size as needed
```

```python
# Generate random filters
random_filters = [np.random.rand(filter_size, filter_size) for _ in range(num_filters)]
```

```python
# Convolve the image with each filter
feature_maps = [scipy.signal.convolve2d(np.array(grayscale_image), filt, mode='same') for filt in random_filters]
```

```python
# Plot each resulting image alongside the corresponding filter
plt.figure(figsize=(15, 8))

for i in range(num_filters):
    plt.subplot(2, num_filters, i + 1)
    plt.imshow(random_filters[i], cmap="gray")
    plt.title(f"Filter {i + 1}")
    plt.axis("off")

    plt.subplot(2, num_filters, i + 1 + num_filters)
    plt.imshow(feature_maps[i], cmap="gray")
    plt.title(f"Feature Map {i + 1}")
    plt.axis("off")

plt.show()
```
